## Churn rate

Не все бронирования завершились успешно (is_canceled), поэтому можно посчитать, сколько клиентов было потеряно в процессе.

Churn rate (отток, коэффициент оттока) – это процент подписчиков (например, на push-уведомления от сайта), которые отписались от канала коммуникации, отказались от услуг сервиса в течение определенного периода времени. Иными словами, представляет собой отношение количества ушедших пользователей к общему количеству пользователей, выраженное в процентах.

В нашем случае Churn Rate - это процент клиентов, которые отменили бронирование. Давайте посмотрим, как эта метрика связана с наличием детей у клиентов!

Создадим переменную has_kids, которая принимает значение True, если клиент при бронировании указал хотя бы одного ребенка (total_kids), в противном случае – False.


```python
bookings['has_kids'] = (bookings['total_kids']>0)
```

```python
#Churn rate для клиентов с детьми
round(bookings[((bookings['has_kids']==True) & (bookings['is_canceled']==True))].shape[0]/bookings[(bookings['has_kids']==True)].shape[0]*100,2)
```

```python
#Churn rate для клиентов без детей
round(bookings[((bookings['has_kids']==False) & (bookings['is_canceled']==True))].shape[0]/bookings[(bookings['has_kids']==False)].shape[0]*100,2)
```

Вывод: Можно заметить, что бОльший отток клиентов произошел в группе, запланировавших отпуск без детей.

Так метрика Churn rate составила 34.92 в группе клиентов с детьми, и 37.22 в группе без детей.
